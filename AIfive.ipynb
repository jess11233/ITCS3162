```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
s = 0.2
delta = 0.05
T = 50  # number of periods

# Arrays to store values
k = np.zeros(T+1)
y = np.zeros(T+1)

# Initial value
k[0] = 20

# Simulation
for t in range(T):
    y[t] = k[t]**(1/3)
    k[t+1] = k[t] + s * y[t] - delta * k[t]

# Last y value
y[T] = k[T]**(1/3)

# Plot
plt.figure(figsize=(10, 5))

plt.plot(k, label='k_t (capital)')
plt.plot(y, label='y_t (output)')

plt.axhline(8, linestyle='--', color='gray', label='k* = 8')

plt.title('Convergence from High Capital')
plt.xlabel('Time (t)')
plt.ylabel('Levels')
plt.legend()
plt.grid(True)

plt.show()
```
